```python
@patch
def summary(self:type(converter.cc)):
    coco = self.gt
    catid2name = {cat['id']:cat['name'] for cat in coco.cats.values()}
    data = dict()
    for ann in coco.anns.values():
        catname = catid2name[ann['category_id']]
        if not catname in data: data[catname] = 0
        data[catname] += 1
    df_ann = pd.DataFrame.from_dict(data, orient='index', columns=['count'])
    print(df_ann)
    data = dict()
    for img in coco.imgs.values():
        print(img)
        break
```

```python
converter.cc.summary()
```
